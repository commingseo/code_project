```python
#https://www.upgrad.com/blog/data-preprocessing-in-machine-learning/ 전처리 과정
#https://stackoverflow.com/questions/39684548/convert-the-string-2-90k-to-2900-or-5-2m-to-5200000-in-pandas-dataframe 단위변환 참고
#DF 단위통일


def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    elif 'K' in x:
        if len(x) > 0.0001:
            return float(x.replace('K', '')) * 1000
      
    elif 'M' in x:
        if len(x) > 0.0001:
            return float(x.replace('M', '')) * 1000000
        
    elif '' in x:
      if len(x) > 0.0001:
        return float(x) * 1000000
    
```

```python
df['NA_Sales'] = df['NA_Sales'].apply(value_to_float)
df['EU_Sales'] = df['EU_Sales'].apply(value_to_float)
df['JP_Sales'] = df['JP_Sales'].apply(value_to_float)
df['Other_Sales'] = df['Other_Sales'].apply(value_to_float)
```
